```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
```

```python
data = pd.read_csv("Train_file(7).csv",header=None)
data.head()
```

```python
def pca(X,k):
     
    X_centered = X - np.mean(X, axis=0)
    covariance_matrix = np.cov(np.transpose(X_centered))
    print(covariance_matrix.shape)

    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:,sorted_indices]
    k_eigenvectors = sorted_eigenvectors[:,:k]
    print(np.linalg.norm(sorted_eigenvectors[:,1]))
    print(k_eigenvectors.shape)
    reduced_dataset = np.dot(X_centered, k_eigenvectors)
    
    print(reduced_dataset.shape)
    return reduced_dataset,train_mean
```

```python
def pca_test(X,k,train_mean):
    print(train_mean)
    X_centered = X - train_mean
    covariance_matrix = np.cov(np.transpose(X_centered))
    print(covariance_matrix.shape)

    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:,sorted_indices]
    k_eigenvectors = sorted_eigenvectors[:,:k]
    print(k_eigenvectors.shape)
    reduced_dataset = np.dot(X_centered, k_eigenvectors)
    
    print(reduced_dataset.shape)
    return reduced_dataset 
```

```python
data = np.array(data)
train_data = data[:,:-1]
train_labels = data[:,-1]
data_new,train_mean  = pca(train_data,20)
```

```python
model1 = keras.Sequential(
    [
        layers.Dense(10, activation="relu", name="layer1"),
        layers.Dense(1, activation="sigmoid")
    ]
)
```

```python
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss=tf.keras.losses.BinaryCrossentropy(),metrics=["acc"])
```

```python
model1.fit(data_new, train_labels, epochs=10)
```

```python
model2 = keras.Sequential(
    [
        layers.Dense(40, activation="relu", name="layer1"),
        layers.Dense(20, activation="relu", name="layer2"),
        layers.Dense(10, activation="relu", name="layer3"),
        layers.Dense(1, activation="sigmoid")
    ]
)
```

```python
model2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss=tf.keras.losses.BinaryCrossentropy(),metrics=["acc"])
```

```python
model2.fit(train_data, train_labels, epochs=12)
```
