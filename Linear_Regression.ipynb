# **Linear Regression Implementation - Gradient Descent for single feature**

```python
import numpy as np
import matplotlib.pyplot as plt
import copy, math
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
```

```python
def compute_cost(x, y, w, b):
  m = x_train.shape[0]
  cost_sum = 0
  for i in range(m):
     f_wb = w*x[i] + b
     cost = (f_wb - y[i])**2
     cost_sum = cost_sum + cost
  total_cost = (1/(2*m))*cost_sum
  return total_cost 
```

```python
def compute_gradient(x,y,w,b):
  m = x_train.shape[0] 
  dj_dw = 0
  dj_db = 0
  for i in range(m):
    f_wb = w*x[i] + b
    dj_dw_i = (f_wb - y[i])*x[i]
    dj_db_i = f_wb - y[i]
    dj_dw = dj_dw + dj_dw_i
    dj_db = dj_db + dj_db_i
  dj_dw= dj_dw/m
  dj_db= dj_db/m

  return dj_dw, dj_db 
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,cost_function,gradient_function):
  w = w_in
  b = b_in 
  i = 0
  while(i < 1000):
    dj_dw,dj_db = gradient_function(x,y,w,b)
    w = w - alpha*dj_dw
    b = b - alpha*dj_db
    i = i + 1
    print(f"Iteration {i:4}",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
  return w,b
  
```

```python
w = 0
b = 0
alpha = 10**(-2)
w_final,b_final = gradient_descent(x_train,y_train,w,b,alpha,compute_cost,compute_gradient)
print(f'w_final:{w_final} b_final:{b_final}')
```

```python
prediction = w_final*1.5 + b_final
print(f"predicted value is :: {prediction}")
```

# **Gradient Descent for Multiple feature Linear Regression** 

```python
def compute_gradient_mltivar(x,y,w,b):
    m,n = x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.
    for i in range(m):
       err = (np.dot(x[i],w) + b) - y[i]
       for j in range(n):
         dj_dw[j] = dj_dw[j] + err*x[i,j]
       dj_db = dj_db + err
    dj_dw= dj_dw/m
    dj_db= dj_db/m
    return dj_dw , dj_db 
```

```python
X_train = np.array([[2104,5,1,45],[1416,3,2,40],[852,2,1,35]])
Y_train = np.array([460,232,178])
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,cost_function,gradient_function):
  w = copy.deepcopy(w_in)
  b = b_in 
  i = 0
  while(i < 1000):
    dj_dw,dj_db = gradient_function(x,y,w,b)
    w = w - alpha*dj_dw
    b = b - alpha*dj_db
    i = i + 1
  return w,b
```

```python
w_init = np.array([0,0,0,0])
b_init = 0.
alpha = 5.0e-7
w_final,b_final = gradient_descent(X_train,Y_train,w_init,b_init,alpha,compute_cost,compute_gradient_mltivar)
print(f'w_final:{w_final} b_final:{b_final:0.2f}')
```

```python
m,n = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {Y_train[i]}")
```
